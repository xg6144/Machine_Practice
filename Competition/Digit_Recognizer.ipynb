```python
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('../digit-recognizer/train.csv', dtype=np.float32)
test_data = pd.read_csv('../digit-recognizer/test.csv')
labels = train_data.pop('label').astype('int64') #pop을 사용해 label을 제거
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_target = labels.to_numpy()
```

```python
train_target
```

```python
train_data = train_data.to_numpy() / 255.0
```

```python
train_data = train_data.reshape(-1, 28, 28, 1)
```

```python
train_target = train_target.reshape(-1, 1)
```

```python
train_target.shape
```

```python
train_x, val_x, train_y, val_y = train_test_split(
    train_data, train_target, test_size = 0.2
)
```

```python
train_x.shape
```

```python
val_x.shape
```

```python
model = keras.Sequential()
```

```python
model.add(layers.Flatten(input_shape=(28,28)))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
model.fit(train_x, train_y, epochs=6)
```

```python
model.evaluate(val_x, val_y)
```

```python
test_input = test_data.to_numpy()
```
