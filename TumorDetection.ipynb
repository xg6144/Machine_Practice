```python
import cv2
import numpy as np
import sys
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
```

```python
model = load_model('modelv1.h5')
```

```python
cap = cv2.VideoCapture('./Normal1.jpeg')

if not cap.isOpened():
    print('Video Load Failed')
    sys.exit()

if cap.isOpened():
    ret, img = cap.read()

    if not ret:
        print('Image Load Failed')
        sys.exit()

    tumor_img = cv2.resize(img, (224, 224))
    x = img_to_array(tumor_img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    tumor_no, tumor_yes = model.predict(x).squeeze()

    print('Yes : ', (tumor_yes * 100),'\n', 'No : ', (tumor_no * 100))

    plt.imshow(img)
    plt.show()
cap.release()
```
